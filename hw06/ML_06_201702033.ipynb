```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from sklearn.metrics.cluster import adjusted_rand_score
```

```python
df = pd.read_csv('iris.csv')
data = np.array(df, dtype='float32')
```

```python
loc2 = 0
loc3 = 0
for i in range(len(data)):
    if data[i,4] == 2 and loc2 == 0:
        loc2 = i
    if data[i,4] == 3 and loc3 == 0:
        loc3 = i
```

```python
plt.scatter(data[:loc2,0],data[:loc2,1], color='r')
plt.scatter(data[loc2:loc3,0],data[loc2:loc3,1], color='b')
plt.scatter(data[loc3:,0],data[loc3:,1], color='g')
plt.show()
```

```python
model_3 = KMeans(n_clusters=3, random_state=0)
model_3.fit(data)
label_3 = model_3.predict(data)
x = data[:,0]
y = data[:,1]
plt.scatter(x,y,c=label_3)
plt.show()
```

```python
model_2 = KMeans(n_clusters=2, random_state=0)
model_4 = KMeans(n_clusters=4, random_state=0)
model_5 = KMeans(n_clusters=5, random_state=0)
model_6 = KMeans(n_clusters=6, random_state=0)
```

```python
model_2.fit(data)
model_4.fit(data)
model_5.fit(data)
model_6.fit(data)
```

```python
label_2 = model_2.predict(data)
label_4 = model_4.predict(data)
label_5 = model_5.predict(data)
label_6 = model_6.predict(data)
```

```python
score2 = adjusted_rand_score(data[:,4], label_2)
score3 = adjusted_rand_score(data[:,4], label_3)
score4 = adjusted_rand_score(data[:,4], label_4)
score5 = adjusted_rand_score(data[:,4], label_5)
score6 = adjusted_rand_score(data[:,4], label_6)
```

```python
print("n clusters가 2일 때:  {:.2f}".format(score2))
print("n clusters가 3일 때:  {:.2f}".format(score3))
print("n clusters가 4일 때:  {:.2f}".format(score4))
print("n clusters가 5일 때:  {:.2f}".format(score5))
print("n clusters가 6일 때:  {:.2f}".format(score6))
```
